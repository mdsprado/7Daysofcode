```python
# lendo a tabela que está os arquivos csv de emprestimo para listar todos eles
```

```python
arquivos = [arq for arq in pd.read_html(url_emprestimo)[0]['Name'].unique() if "emprestimo" in arq]
arquivos
```

```python
url_raw = "https://raw.githubusercontent.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/main/Dia_1-Importando_dados/Datasets/dados_emprestimos/"
```

```python
for p, arq in enumerate(arquivos):
    if p==0:
        df=pd.read_csv(url_raw+arq)
    else:
        df = pd.concat([df,pd.read_csv(url_raw+arq)])
df
```

# Dia 2

> inserir coluna classe geral da CDU
- 000 a 099: Generalidades. Ciência e conhecimento.
- 100 a 199: Filosofia e psicologia.
- 200 a 299: Religião.
- 300 a 399: Ciências sociais.
- 400 a 499: Classe vaga. Provisoriamente não ocupada.
- 500 a 599: Matemática e ciências naturais.
- 600 a 699: Ciências aplicadas.
- 700 a 799: Belas artes.
- 800 a 899: Linguagem. Língua. Linguística.
- 900 a 999: Geografia. Biografia. História."

> excluir coluna registro_sistema

> transforma "matricula_ou_siape" em string


```python
def classe_CDU(loc):
    if loc in [n for n in range(0,100)]:
        return "Generalidades. Ciência e conhecimento"
    elif loc in [n for n in range(100,200)]:
        return "Filosofia e psicologia"
    elif loc in [n for n in range(200,300)]:
        return "Religião"
    elif loc in [n for n in range(300,400)]:
        return "Ciências sociais"
    elif loc in [n for n in range(400,500)]:
        return "Classe vaga. Provisoriamente não ocupada"
    elif loc in [n for n in range(500,600)]:
        return "Matemática e ciências naturais"
    elif loc in [n for n in range(600,700)]:
        return "Ciências aplicadas"
    elif loc in [n for n in range(700,800)]:
        return "Belas artes"
    elif loc in [n for n in range(800,900)]:
        return "Linguagem. Língua. Linguística"
    elif loc in [n for n in range(900,1000)]:
        return "Geografia. Biografia. História"
    
```

```python
df['classe_CDU'] = df['localizacao'].apply(classe_CDU)
```

```python
df.sample(5)
```

```python
df.groupby('classe_CDU').count()
```

```python
df.drop("registro_sistema",axis=1)
```

```python
#converti em inteiro primeiro e depois em string, pois estava vindo o .0 no final, quando convertia direto de float para string
df['matricula_ou_siape']=df['matricula_ou_siape'].astype(int)
df['matricula_ou_siape']=df['matricula_ou_siape'].astype(str)
```
